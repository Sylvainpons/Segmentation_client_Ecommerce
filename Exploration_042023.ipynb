```python
# Convertir la colonne order_purchase_timestamp en datetime
df_order_list['order_purchase_timestamp'] = pd.to_datetime(df_order_list['order_purchase_timestamp'])

# Calculer la date la plus rÃ©cente
latest_date = df_order_list['order_purchase_timestamp'].max()

# Calculer la rÃ©cence, la frÃ©quence et le montant pour chaque client
# regroupe les commandes par client.
rfm = df_order_list.groupby('customer_id').agg({
    #calcule la diffÃ©rence de temps entre la date la plus rÃ©cente (latest_date) et la date d'achat la plus rÃ©cente pour chaque client (x.max()).
    'order_purchase_timestamp': lambda x: (latest_date - x.max()).days,
    'order_id': 'nunique',
}).reset_index()

# Renommer les colonnes
rfm.rename(columns={'order_purchase_timestamp': 'recency',
                     'order_id': 'frequency'}, inplace=True)

# Calculer le montant pour chaque client
monetary = df_order_item.groupby('order_id').agg({'price': 'sum'}).reset_index()
#Je merge les colonnes 'customer_id' et 'order_id' du dataframe 'df_order_list' au dataframe 'rfm' fusionner les donnÃ©es de deux dataframes sur'customer_id'
rfm = pd.merge(rfm, df_order_list[['customer_id', 'order_id']], on='customer_id')
rfm = pd.merge(rfm, monetary, on='order_id')
#Je drop la colonne order_id car j'ai fusionnÃ© les rfm et monetary sur la colonne order_id est elle est plus nÃ©cessaire maintenant
rfm.drop(['order_id'], axis=1, inplace=True)
rfm = rfm.groupby('customer_id').agg({'recency': 'min',
                                      'frequency': 'min',
                                      'price': 'sum'}).reset_index()
rfm.rename(columns={'price': 'monetary_value'}, inplace=True)

# Afficher le rÃ©sultat
print(rfm.head())
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Standardisation des donnÃ©es
scaler = StandardScaler()
rfm_std = scaler.fit_transform(rfm[['recency', 'frequency', 'monetary_value']])

# Initialisation du modÃ¨le KMeans avec le nombre de clusters souhaitÃ©
kmeans = KMeans(n_clusters=5, random_state=1, n_init=10).fit(rfm_std)

# Visualisation des clusters
plt.scatter(rfm_std['recency'], rfm_std['monetary_value'], c=kmeans.labels_)
plt.title("KMeans Clusters")
plt.show()

```

```python
from sklearn.cluster import AgglomerativeClustering
#Permet le calcul ari entre deux ensembles d'Ã©tiquettes de cluster
from sklearn.metrics import adjusted_rand_score

# CrÃ©ation d'une liste vide pour stocker le ARI scores
ari_scores = []

# CrÃ©ation du cluster initial pour le test de maintenance last_order_date est retirÃ© car par utilisÃ© pour le cluster initial
cluster_a = AgglomerativeClustering(n_clusters=9, linkage='ward', metric='euclidean').fit(bloc_a.drop('last_order_date', axis=1))
#RÃ©cupÃ¨re les Ã©tiquettes de cluster_a
cluster_a_labels = cluster_a.labels_
#Pour chaque bloc de donnÃ©es suivant
for i in range(1, 7):
    
    # SÃ©lectionne le bloc de donnÃ©es correspondant aux commandes passÃ©es au cours des i derniers mois moins actuel - i mois et date actuel + i 1 mois
    block_b = df_maintenance.loc[(df_maintenance['last_order_date'] <= min_six) & (df_maintenance['last_order_date'] > min_six - pd.DateOffset(months=i)), :]
    
    #  ConcatÃ¨ne les deux blocs de donnÃ©es pour former un nouveau bloc de donnÃ©es donnÃ©e tous les mois prÃ©cedent + mois actuel
    block_a1 = pd.concat([bloc_a, block_b])
    # CrÃ©e un nouveau cluster sur le nouveau bloc de donnÃ©es bloc_a1. Les paramÃ¨tres sont les mÃªmes que pour le premier cluster
    cluster_a1 = AgglomerativeClustering(n_clusters=9, linkage='ward', metric='euclidean').fit(block_a1.drop('last_order_date', axis=1))
    
    # RÃ©cupÃ¨re les Ã©tiquettes de cluster_a1
    cluster_a1_labels = cluster_a1.labels_
  
    # Calcule le score ARI entre l'ensemble d'Ã©tiquettes de cluster du premier bloc et du nouveau bloc de donnÃ©es.
    ari_score = adjusted_rand_score(cluster_a_labels, cluster_a1_labels[:len(cluster_a_labels)])
    
    # Ajouter les score Ã  la list ari_score
    ari_scores.append(ari_score)
    
    # DÃ©finit le nouveau cluster comme le cluster original pour la prochaine itÃ©ration
    cluster_a = cluster_a1
    cluster_a_labels = cluster_a.labels_
    block_a = block_a1
    
# Afficher dans un graphique le score ari
plt.plot(range(1, 7), ari_scores)
plt.xlabel('Months')
plt.ylabel('ARI Score')
plt.show()
```
